# Evaluate the results for cause and effect two sentences

Problem: there are many ways to give a correct answer and it's not easy to find all of them. 
Our rules:
- If only the correct sentence is repeated it counts as a correct answer
- If any other sentence is added it counts as an incorrect answer
- If the model says "the first sentence" or "the second sentence" it counts as true if it refers to the correct sentence
- Stuff like "the sentence with the child" does not count because it is too hard to evaluate automatically, even if a human could understand it as referring to the correct answer.
